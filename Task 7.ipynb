```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 

warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("data.csv")
```

```python
df= df.drop("id", axis=1)
```

```python
df=df.drop("Unnamed: 32", axis=1)
```

```python
df["diagnosis"].value_counts()
```

```python
df["diagnosis"]=df["diagnosis"].map({"B":0, "M":1})
```

```python
df
```

```python
x=df.drop("diagnosis", axis=1)
y=df["diagnosis"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=1)

```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_train_scaled=scalar.fit_transform(x_train)
x_test_scaled=scalar.transform(x_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score
svc_linear=SVC(kernel='linear',C=1)
svc_linear.fit(x_train_scaled, y_train)
```

```python
y_pred=svc_linear.predict(x_test)
```

```python
svc_rbf=SVC(kernel='rbf',C=1, gamma='scale')
svc_rbf.fit(x_train_scaled,y_train)
```

```python
y_pred_rbf=svc_rbf.predict(x_test)
```

```python
print("Linear Kernel Accuracy:", accuracy_score(y_test, y_pred))
print("RBF Kernel Accuracy:", accuracy_score(y_test, y_pred_rbf))
```

```python
import numpy as np 
#Visualize Decision Boundary (2D projection)
# Select first two features for plotting
x_vis = x.iloc[:, :2]
x_vis_scaled=StandardScaler().fit_transform(x_vis)
```

```python
x_train_v ,x_test_v , y_train_v, y_test_v= train_test_split(x_vis_scaled,y, test_size=0.2, random_state=1)
```

```python
#fit SVM on 2D data
svc_vis=SVC(kernel='linear',C=1)
svc_vis.fit(x_train_v, y_train_v)  
```

```python
# Meshgrid for contour plot
xx, yy = np.meshgrid(
    np.linspace(x_vis_scaled[:,0].min()-1, x_vis_scaled[:,0].max()+1, 500),
    np.linspace(x_vis_scaled[:,1].min()-1, x_vis_scaled[:,1].max()+1, 500)
)
Z = svc_vis.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot
plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.2, cmap=plt.cm.coolwarm)
plt.scatter(x_train_v[:,0], x_train_v[:,1], c=y_train_v, cmap=plt.cm.coolwarm, edgecolors='k')
plt.title("SVM Linear Kernel Decision Boundary (2D projection)")
plt.show()
```

```python
#Hyperparameter Tuning (C, gamma)
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 0.01, 0.1, 1]
}

grid = GridSearchCV(SVC(kernel='rbf'), param_grid, cv=5)
grid.fit(x_train, y_train)

print("Best parameters:", grid.best_params_)
print("Best cross-validation score:", grid.best_score_)

# Evaluate on test set
y_pred_grid = grid.predict(x_test)
print("Test accuracy with tuned parameters:", accuracy_score(y_test, y_pred_grid))
```

```python
#Cross-validation Performance
from sklearn.model_selection import cross_val_score

# Use best estimator from GridSearch
best_svc = grid.best_estimator_

cv_scores = cross_val_score(best_svc, x_train_scaled, y_train, cv=5)

print("Cross-validation scores:", cv_scores)
print("Mean CV accuracy:", np.mean(cv_scores))
```
